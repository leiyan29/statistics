# Random sampling

### Some definitions

1. stratified sampling: random samples based on stratas, i.e. splitting population by age groups and randomly pulling samples from each strata

2. sample biais: a sample misrepresenting the population, i.e. polling one city when you want to understand the political leanings of the entire country

3. data snooping: find something inteAresting in data without looking through all of your data before modeling

4. vast search effect: to find outliers

5. data distribution: frequency distribution of individual values in a dataset

6. central limit theorem: the means drawn from multiple samples from data ressemble a bell-shaped normal curve
